# **Feature Selection Advanced House Price Prediction**
The main aim of this project is to predict the house price based on various features which we will discuss as we go ahead

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('Data/X_train.csv')
test= pd.read_csv("data/X_test.csv")
train.head()
```

```python
## Capture the dependent feature
y_train=train[['SalePrice']]
y_test=test[['SalePrice']]
```

```python

## drop dependent feature from dataset
x_train=train.drop(['Id','SalePrice'],axis=1)
x_test=test.drop(['Id','SalePrice'],axis=1)
```

```python
# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) 
feature_sel_model.fit(x_train, y_train)
```

```python
selected_feature = x_train.columns[feature_sel_model.get_support()]
print(selected_feature)
```

```python
print('total features: {}'.format((x_train.shape[1])))
print('selected features: {}'.format(len(selected_feature)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python

x_train=x_train[selected_feature]
x_train.head()
```

```python
x_test = x_test[selected_feature]
x_test.head()
```

```python
x_train.to_csv('Data/2.x_train.csv',index=False)
```

```python
x_test.to_csv('Data/2.x_test.csv',index=False)
y_train.to_csv('Data/2.y_train.csv',index=False)
y_test.to_csv('Data/2.y_test.csv',index=False)
```
