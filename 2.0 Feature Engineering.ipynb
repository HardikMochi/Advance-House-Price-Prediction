# **Advanced Housing Prices- Feature Engineering**

The main aim of this project is to predict the house price based on various features which we will discuss as we go ahead


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
train=pd.read_csv('Data/train.csv')
train.head()
```

```python
## Always remember there way always be a chance of data leakage so we need to split the data first and then apply feature
## Engineering
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(train,train['SalePrice'],test_size=0.1,random_state=0)
```

```python
X_train.shape, X_test.shape
```

## **Missing Values**

```python

features_nan=[feature for feature in train.columns if train[feature].isnull().sum()>0 and train[feature].dtypes=='object']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(train[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

X_train=replace_cat_feature(X_train,features_nan)
X_test=replace_cat_feature(X_test,features_nan)

X_train[features_nan].isnull().sum()
```

```python
X_train.shape, X_test.shape
```

```python
# now lets check missing value for numerical data
numerical_with_nan = [feature for feature in train.columns if train[feature].isnull().sum()>0 and train[feature].dtypes!='object']
for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(train[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value_t=X_train[feature].median()
    median_value_v=X_test[feature].median()
    
    ## create a new feature to capture nan values
    X_train[feature+'nan']=np.where(X_train[feature].isnull(),1,0)
    X_train[feature].fillna(median_value_t,inplace=True)
    X_test[feature+'nan']=np.where(X_test[feature].isnull(),1,0)
    X_test[feature].fillna(median_value_v,inplace=True)
    
X_train[numerical_with_nan].isnull().sum()
```

```python
X_train.shape, X_test.shape
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    X_train[feature]=X_train['YrSold']-X_train[feature]
    X_test[feature]=X_test['YrSold']-X_test[feature]
```

## **Numerical Variables**

Numerical values has some skewnes

```python

num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    X_train[feature]=np.log(X_train[feature])
    X_test[feature]=np.log(X_test[feature])
```

```python
X_train.head()
```

## **Transforming Categorical Data By Using Target Mean Encoding**

```python
categorical_features=[feature for feature in train.columns if train[feature].dtype=='O']
print(categorical_features)
```

```python
for feature in categorical_features:
    order = X_train.groupby(feature)['SalePrice'].median().sort_values().index
    order = {k:i for i,k in enumerate(order,0)}
    X_train[feature] = X_train[feature].map(order)
    X_test[feature] = X_test[feature].map(order)
  
    
```

## **Feature Scaling**

```python
feature_scale=[feature for feature in X_train.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train[feature_scale])
```

```python
#transform the train and test set, and add on the Id and SalePrice variables
X_train = pd.concat([X_train[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_train[feature_scale]), columns=feature_scale)],
                    axis=1)
X_test = pd.concat([X_test[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X_test[feature_scale]), columns=feature_scale)],
                    axis=1)
```
